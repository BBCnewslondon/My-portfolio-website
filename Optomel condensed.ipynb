```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import integrate

def gaussian(x, height, mean, sigma):
    return height * np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))

def fit_and_plot_histogram(data, threshold, channel_color, bins=256, title='Histogram with Gaussian Fit'):
    """
    Fit a Gaussian to the histogram of the data and plot both.

    Parameters:
    - data: The data to be histogrammed.
    - threshold: Values below this threshold will be ignored.
    - channel_color: Color of the histogram and Gaussian fit, e.g., 'red', 'green', 'blue'.
    - bins: Number of bins for the histogram.
    - title: Title of the plot.
    """
    # Filter data to include only values greater than the threshold
    filtered_data = data[data > threshold]
    
    # Extract histogram data with specified number of bins
    frequencies, bin_edges = np.histogram(filtered_data, bins=bins, range=(threshold, 256))
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    
    # Use curve_fit to fit the Gaussian function to the filtered data
    popt, _ = curve_fit(gaussian, bin_centers, frequencies, p0=[max(frequencies), np.mean(filtered_data), np.std(filtered_data)])
    
    # Plot the histogram and the fitted curve
    plt.figure()
    plt.hist(filtered_data, bins=bins, color=channel_color, alpha=0.5, label='Histogram')
    plt.plot(bin_centers, gaussian(bin_centers, *popt), color=channel_color, linewidth=2, label='Fitted Gaussian')
    plt.title(title)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()


    # Calculate expected frequencies using the fitted Gaussian function
    expected_frequencies = gaussian(bin_centers, *popt) * np.diff(bin_edges) * len(filtered_data)
    
    # Normalize expected frequencies to match the sum of observed frequencies
    expected_frequencies *= np.sum(frequencies) / np.sum(expected_frequencies)
    
    # Calculate RMSE
    rmse = np.sqrt(np.mean((frequencies - expected_frequencies) ** 2))
    
    # Calculate MAE
    mae = np.mean(np.abs(frequencies - expected_frequencies))
    
    # Calculate R-squared
    ss_res = np.sum((frequencies - expected_frequencies) ** 2)
    ss_tot = np.sum((frequencies - np.mean(frequencies)) ** 2)
    r_squared = 1 - (ss_res / ss_tot)
    
    # Extract mean, standard deviation, and calculate FWHM from the fitted parameters
    mean = popt[1]
    std_dev = popt[2]
    fwhm = 2 * np.sqrt(2 * np.log(2)) * std_dev
    #showing the estimated equation of the gaussian
    print(f"Estimated Gaussian: height = {popt[0]:.2f}, mean = {popt[1]:.2f}, sigma = {popt[2]:.2f}")
    
    
    # Print the RMSE, MAE, R-squared, mean, standard deviation, and FWHM values
    print(f"RMSE: {rmse}, MAE: {mae}, R-squared: {r_squared}, Mean: {mean}, Standard Deviation: {std_dev}, FWHM: {fwhm}")
    #saving the values in a text file
    with open('fit_results.txt', 'a') as f:
        f.write(f"for Sample {title}\n RMSE: {rmse}, MAE: {mae}, R-squared: {r_squared}, Mean: {mean}, Standard Deviation: {std_dev}, FWHM: {fwhm}\n")
        
    #finding the integral of the fitted gaussian
    integral, _ = integrate.quad(gaussian, 0, 256, args=(popt[0], popt[1], popt[2]))
    print(f"Integral of the fitted Gaussian: {integral}")
    
    #appending summary of the results in a text file
    with open('fit_results.txt', 'a') as f:
        f.write(f"Integral of the fitted Gaussian: {integral}\n")
        
        
        
        
        
    
    
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import numpy as np

# Load the data
sheet_name = 'Summary Results'
summary = pd.read_excel(r'C:\Users\singh\Downloads\OneDrive_2024-06-18\Optomel Armaan (Sharing Folder)\200624 (1).xlsx', sheet_name=sheet_name)

# Drop unnecessary columns and rows with missing values
summary = summary.drop(columns=['Unnamed: 12'])
summary.dropna(inplace=True)

# Assuming 'Central Pos' is the target variable and the rest are features
target_column = 'Central Pos'
bandwidth_column = 'Bandwith'  # Assuming 'Bandwidth' is the column name for bandwidth data
features = summary.drop(columns=[target_column, bandwidth_column, 'Sample', 'Channel'])
target = summary[target_column]
bandwidth = summary[bandwidth_column]

# Convert non-numeric values to numeric, handle '<90' as an example
features = features.apply(pd.to_numeric, errors='coerce')

# Handle missing values (e.g., fill with mean of the column)
features = features.fillna(features.mean())

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test, bandwidth_train, bandwidth_test = train_test_split(
    features_scaled, target, bandwidth, test_size=0.2, random_state=40)

# Initialize the model
model = GradientBoostingRegressor()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')

# Calculate the average error over multiple examples and check if within bandwidth
errors = []
within_bandwidth_count = 0
num_examples = len(X_test)  # Use all test examples

for i in range(num_examples):
    example_data = X_test[i].reshape(1, -1)
    actual_central_position = y_test.iloc[i]
    predicted_central_position = model.predict(example_data)
    error = abs(actual_central_position - predicted_central_position[0])
    errors.append(error)
    
    # Check if the error is within the bandwidth
    if error <= bandwidth_test.iloc[i]:
        within_bandwidth_count += 1
    
    print(f'Example {i+1}:')
    print(f'  Actual Central Position: {actual_central_position}')
    print(f'  Predicted Central Position: {predicted_central_position[0]}')
    print(f'  Error: {error}')
    print(f'  Bandwidth: {bandwidth_test.iloc[i]}')
    print(f'  Within Bandwidth: {error <= bandwidth_test.iloc[i]}')

average_error = np.mean(errors)
within_bandwidth_percentage = (within_bandwidth_count / num_examples) * 100

print(f'Average Error: {average_error}')
print(f'Percentage of Errors Within Bandwidth: {within_bandwidth_percentage}%')
```
